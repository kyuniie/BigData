```python
T = []
SMILES = []
for i in range(0, len(F)):
    CHEM = F[i][2]
    for j in range(0, len(G)):
        CHEM2 = G[j][13]
        if CHEM == CHEM2:
            T.append([F[i][2],G[j][:10]])
            SMILES.append(F[i][0])
```
